```python
#Criando nosso DataFrame
df = pd.read_excel("dataset/AdventureWorks.xlsx")
```

```python
#Qual a receita total?
df["Valor Venda"].sum()
```

```python
# Custo total do produto
df['Custo Total'] = df['Custo Unitário'] * df['Quantidade']
df.head()
```

```python
# Lucro bruto da venda
df['Lucro Bruto'] = df['Valor Venda'] - df['Custo Total']
df.head()
```

```python
#Criando uma coluna com total de dias para enviar o produto
df["Tempo_envio"] = df["Data Envio"] - df["Data Venda"]
df.head(1)
```

```python
# Extraindo apenas os dias
df["Tempo_envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
df.head(1)
```

```python
# Verificando o tipo da coluna Tempo_envio
df["Tempo_envio"].dtype
```

```python
df.groupby("Marca")["Tempo_envio"].mean()
```

```python
# Vamos agrupar por ano e marca
df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro Bruto"].sum()
```

```python
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro Bruto"].sum().reset_index()
lucro_ano
```

```python
#Qual o total de produtos vendidos?
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=False)
```

```python
#Grafico total de produtos vendidos
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(title="Total Produtos Vendidos")
plt.xlabel("Total")
plt.ylabel("Produto");
```

```python
df.groupby(df["Data Venda"].dt.year)["Lucro Bruto"].sum().plot.bar(title="Lucro Bruto x ano")
plt.xlabel("Ano")
plt.ylabel("Receita");
```

```python
df.groupby(df["Data Venda"].dt.year)["Lucro Bruto"].sum()
```

```python
#Selecionando apenas as vendas de 2009
df_2008 = df[df["Data Venda"].dt.year == 2008]
#Selecionando apenas as vendas de 2009
df_2009 = df[df["Data Venda"].dt.year == 2009]
```

```python
df_2008.head()
```

```python
df_2009.head()
```

```python
df_2008.groupby(df_2008["Data Venda"].dt.month)["Lucro Bruto"].sum().plot(color='red', label='Ano 2008')
df_2009.groupby(df_2009["Data Venda"].dt.month)["Lucro Bruto"].sum().plot(color='green', label='Ano 2009')
plt.title("Lucro x Mês")
plt.xlabel("Mês")
plt.ylabel("Lucro Bruto")
plt.legend()
plt.show()
```

```python
df_2009.groupby("Marca")["Lucro Bruto"].sum().plot.bar(title="Lucro Bruto x Marca")
plt.xlabel("Marca")
plt.ylabel("Lucro Bruto")
plt.xticks(rotation='horizontal');
```

```python
df_2009.groupby("Classe")["Lucro Bruto"].sum().plot.bar(title="Lucro Bruto x Classe")
plt.xlabel("Classe")
plt.ylabel("Lucro Bruto")
plt.xticks(rotation='horizontal');
```
